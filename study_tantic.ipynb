```python
import pandas as pd
tanic=pd.read_csv('D:/data/titanic_train.csv')
tanic.head()
```

```python
tanic["Age"]=tanic["Age"].fillna(tanic["Age"].mean())
tanic["Embarked"]=tanic["Embarked"].fillna("S")
print(tanic.info())
```

```python
tanic.loc[tanic["Sex"]=='male',"Sex"]=0
tanic.loc[tanic["Sex"]=='female',"Sex"]=1

tanic.loc[tanic["Embarked"]=='S',"Embarked"]=0
tanic.loc[tanic["Embarked"]=='C',"Embarked"]=1
tanic.loc[tanic["Embarked"]=='Q',"Embarked"]=2

tanic
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn import tree
from sklearn import model_selection
pre=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
x_data=tanic[pre]
y_data=tanic["Survived"]

stand=StandardScaler()
x_data=stand.fit_transform(x_data)
```

```python
from sklearn import tree
from sklearn import model_selection
from sklearn.model_selection import GridSearchCV, train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=0)
pre_tree=tree.DecisionTreeClassifier(max_depth=5,min_samples_split=4)
scores=model_selection.cross_val_score(pre_tree,X_train,y_train,cv=5)
print(scores.mean())
```

```python
pre_tree.fit(X_train,y_train)
y_predict=pre_tree.predict(X_test)
print(y_predict)
```

```python
from sklearn import model_selection  
from sklearn.linear_model import LogisticRegression

# 逻辑回归模型
LR = LogisticRegression()
# 计算交叉验证的误差
scores = model_selection.cross_val_score(LR,X_train,y_train, cv=5)
# 求平均
print(scores.mean())
```

```python
from sklearn import neighbors

knn = neighbors.KNeighborsClassifier(21)#邻居数量
# 计算交叉验证的误差
scores = model_selection.cross_val_score(knn,  X_train, y_train, cv=5)
# 求平均
print(scores.mean())
```

```python
pre_tree=tree.DecisionTreeClassifier(max_depth=5,min_samples_leaf=1,min_samples_split=3,max_features=None)
print('决策树：',model_selection.cross_val_score(pre_tree, X_train, y_train, cv=5).mean())

LR = LogisticRegression(C=0.1,penalty='l2')
print('逻辑回归：',model_selection.cross_val_score(LR, X_train, y_train, cv=5).mean())

knn = neighbors.KNeighborsClassifier(n_neighbors=5,weights='uniform',algorithm='ball_tree')
print('Knn：',model_selection.cross_val_score(knn, X_train, y_train, cv=5).mean())
```

```python
from sklearn.ensemble import VotingClassifier
from mlxtend.classifier import StackingClassifier 


sclf2 = VotingClassifier([('LR',LR),('knn',knn),('dtree',pre_tree)])  


scores = model_selection.cross_val_score(sclf2, X_train, y_train, cv=5)

print(scores.mean())
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()

# 定义超参数选项
parameters = {'max_depth': [3, 4, 5, 6, 7, 8, 9, 10],
              'min_samples_split': [2, 3, 4, 5, 6, 7],
              'min_samples_leaf': [1, 2, 3, 4, 5, 6],
              'max_features': ['sqrt', 'log2', None]}

# 定义网格搜索对象
grid_search = GridSearchCV(model, parameters, cv=5)

# 训练模型
grid_search.fit(X_train, y_train)

# 输出最佳超参数组合和对应的得分
print("Best parameters: {}".format(grid_search.best_params_))
print("Best score: {:.2f}".format(grid_search.best_score_))
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,roc_auc_score
k_range = [3, 5, 7, 9, 11, 13, 15]
scores = []
scores_mean=[]
accuracy=[]

for k in k_range:
    knn = neighbors.KNeighborsClassifier(n_neighbors=k,weights='uniform',algorithm='ball_tree')
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy.append(accuracy_score(y_test, y_pred))
    scores.append(roc_auc_score(y_test, y_pred))

plt.plot(k_range, accuracy,label='accuracy',color='red')
plt.plot(k_range, scores,label='scores',color='blue')

plt.plot(k_range,scores)
plt.xlabel('K_range')
plt.ylabel('two assess ')
plt.title('KNN')
plt.legend()
plt.show()
```
